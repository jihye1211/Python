```python
# 예외 처리 구문 : try -except문

for i in range(10):
    try:
        print(10 / i)
    except ZeroDivisionError:
        print("Not divided by 0")
```

```python
for i in range(10):
    try:
        print(10 / i)
    except ZeroDivisionError as e:
        print(e)
        print("Not divided by 0")
```

```python
# 예외 처리 구문 : try-except-else문

for i in range(10):
    try:
        result = 10 / i
    except ZeroDivisionError:
        print("Not divided by 0")
    else:
        print(result)
```

```python
# 예외 처리 구문 : try-except-finally문

try:
    for i in range(1, 10):
        result =10 // i
        print(result)
except ZeroDivisionError:
    print("Not divided by 0")
finally:
    print("종료되었다.")
```

```python
# 파일 읽기 : with문과 함께 사용하기

with open("dream.txt","r") as my_file:
    contents = my_file.read()
    print(type(contents), contents)
```

```python
# 파일 읽기 : 한줄씩 읽어 리스트형으로 반환하기 readlines()함수 사용

with open("dream.txt","r") as my_file:
    content_list = my_file.readlines()      # 파일 전체를 리스트로 반환
    print(type(content_list))               # 자료형 확인
    print(content_list)  
```

```python
# 파일 읽기 : 실행할 때마다 한줄씩 읽어오기

with open("dream.txt", "r") as my_file:
    i = 0
    while 1:
        line = my_file.readline()
        if not line:
            break
        print(str(i)+" === "+ line.replace("\n",""))    # 한 줄씩 값 출력
        i = i + 1
```

```python
# 파일 읽기 : 파일 안 글자의 통계 정보 출력하기

with open("dream.txt", "r") as my_file:
    contents = my_file.read()
    word_list = contents.split(" ")             # 빈칸 기준으로 단어를 분리 리스트
    line_list = contents.split("\n")            # 한 줄씩 분리하여 리스트
    
print(word_list)
print(line_list)

print("총 글자의 수:", len(contents))
print("총 단어의 수:", len(word_list))
print("총 줄의 수:", len(line_list))
```

```python
# 파일 자동 다운로드
# 시간이 많이 걸려 설명한 하고 실행하지 않음
# 리스트만 확인
import urllib.request                           # urllib 모듈 호출
import re

url = "http://www.google.com/googlebooks/uspto-patents-grants-text.html" # url 값 입력
html = urllib.request.urlopen(url)      # url 열기
html_contents = str(html.read().decode("utf8")) # html 파일 읽고, 문자열로 변환

url_list = re.findall(r"(http)(.+)(zip)", html_contents)  # html 파일 읽고, 문자열로 변환
for url in url_list:
    full_url = "".join(url)                               # 출력된 튜플 형태 데이터 문자열로 join
    print(full_url)
    print(full_url[59:])
#    fname, header = urllib.request.urlretrieve(full_url, file_name) # file_name에 다운로드할 파일명 입력한 후, 파일 다운로드
#    위에 코드에 리스트 값 입력하면 파일 다운로드
print ("End Download")
```
