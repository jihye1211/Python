```python
import random
guess_number = random.randint(1,20)   #1~20  
print("  (1~20)")
user_input = int(input())

while(user_input is not guess_number):
    if user_input > guess_number:
        print("  .")
    else:
        print("  .")
    user_input = int(input())
else:
    print(".", "  ", user_input, ".")
```

```python
print("   (1~9)?")
user_input = 1

while(user_input is not 0):
    user_input = int(input())
    if(user_input == 0):
        break
    if not(1 <= user_input <= 9):
        print("  . 1~9 ")
        continue
    else:
        print("", user_input, " .")
        for i in range(1,10):
            result = user_input * i
            print(str(user_input) + "X" + str(i) + "=" + str(result))

print("  .")
```
