```python
import pandas as pd
```

### 서울시 CCTV 현황 통계자료 읽어오기

```python
# 엑셀 파일 및 CSV 파일 읽어오기
# pandas에서는 자동으로 첫 행을 컬럼명으로 인식함
# CCTV_Gu = pd.read_csv('14_Seoul_Gu_CCTV_2013-2018.csv',  encoding='utf-8')
CCTV_Gu = pd.read_excel('14_Seoul_Gu_CCTV_2013-2018.xlsx', sheet_name='Excel_Import_1',  encoding='utf-8')
CCTV_Gu.head()
```

```python
# 컬럼명 확인 : columns 사용
CCTV_Gu.columns
```

```python
CCTV_Gu.columns[0]
```

```python
# 컬럼명을 변경하고자 할때 사용 : rename
CCTV_Gu.rename(columns={CCTV_Gu.columns[0] : '구별'}, inplace=True)
CCTV_Gu.head()
```

### 서울시 인구현황 통계자료 읽어오기

```python
# 서울시 인구현황 엑셀 파일 읽어오기
Seoul_population = pd.read_excel('14_population_Gu.xls',  encoding='utf-8')
Seoul_population.head()
```

```python
# 파일을 읽을때 특정한 컬럼만 가져 오고 싶은 경우 처리
Seoul_population1 = pd.read_excel('14_population_Gu.xls', 
                          header = 2,
                          usecols = 'B, D, G, J, N',
                          encoding='utf-8')
Seoul_population1.head()
```

```python
# 컬럼명을 변경하고자 할때 rename 사용
Seoul_population1.rename(columns={Seoul_population1.columns[0] : '구별', 
                          Seoul_population1.columns[1] : '인구수', 
                          Seoul_population1.columns[2] : '한국인', 
                          Seoul_population1.columns[3] : '외국인', 
                          Seoul_population1.columns[4] : '고령자'}, inplace=True)
Seoul_population1.head()
```

### 서울시 CCTV 데이터 분석하기

```python
CCTV_Gu.head()
```

```python
# 데이터 정렬 사용 : sort_values() 함수 사용
  # by = ['칼럼명'] : 칼럼명을 기준으로 정렬
  # ascending=[True] : 오름차순으로 정렬, Flase : 내림차순으로 정렬
CCTV_Gu.sort_values(by='소계', ascending=True).head(5)
```

```python
CCTV_Gu.sort_values(by='소계', ascending=False).head(5)
```

```python
# "증가율" 마지막 컬럼 추가하고자 하는경우 
CCTV_Gu['증가율'] = (CCTV_Gu['2017년'] + CCTV_Gu['2018년'] + \
                        CCTV_Gu['2016년'] + CCTV_Gu['2015년'] + \
                        CCTV_Gu['2014년']) / CCTV_Gu['2013년이전']  * 100
CCTV_Gu.sort_values(by='증가율', ascending=True).head(5)
```

### 서울시 구별 인구 데이터 분석하기

```python
Seoul_population1.head()
```

```python
# Pandas에서 유일한 값 찾을때 사용 함수: uynique() 
Seoul_population1['구별'].unique()
```

```python
# 데이터가 null인 값을 검색할때 사용 함수 : isnull()
Seoul_population1[Seoul_population1['구별'].isnull()]
```

```python
# 무의민한 데이터 행 삭제 함수 : drop()함수
Seoul_population1.drop([26], inplace=True)
Seoul_population1.head()
```

```python
# 데이터가 null인 값을 검색할때 사용 함수 : isnull()
Seoul_population1[Seoul_population1['구별'].isnull()]
```

```python
# 컬럼 추가 : 외국인비율, 고령자비율
Seoul_population1['외국인비율'] = Seoul_population1['외국인'] / Seoul_population1['인구수'] * 100
Seoul_population1['고령자비율'] = Seoul_population1['고령자'] / Seoul_population1['인구수'] * 100
Seoul_population1.head()
```

```python
# 고령자 비율로 정렬
Seoul_population1.sort_values(by='고령자비율', ascending=False).head(5)
```

### 서울시 CCTV현황과 인구현황 병합하여 분석하기

```python
# Pandas 두개를 하나로 병합 함수 : merge() 함수
# error 발생 : 구별이 맞지 않아서
cctv_pop_merge = pd.merge(CCTV_Gu, Seoul_population1, on='구별')
cctv_pop_merge.head()
```

```python
CCTV_Gu.head()
```

```python
Seoul_population1.head()
```

```python
# 컬럼명을 변경하고자 할때 rename 사용
CCTV_Gu.rename(columns={CCTV_Gu.columns[0] : '구별', }, inplace=True)
CCTV_Gu.head()
```

```python
# CCTV_Gu 에 공백이 있어서 정상으로 merge가 되지 않음. 정규표현식 : 공백이 1개 이상 (\s+)
CCTV_Gu['구별'] = CCTV_Gu['구별'].str.replace(r"(\s+)", "")
CCTV_Gu.head()
```

```python
# Pandas 두개를 하나로 병합 함수 : merge() 함수
cctv_pop_merge = pd.merge(CCTV_Gu, Seoul_population1, on='구별')
cctv_pop_merge.head()
```

```python
# 필요없는 칼럼 삭제 : del 사용
del cctv_pop_merge['2013년이전']
del cctv_pop_merge['2014년']
del cctv_pop_merge['2015년']
del cctv_pop_merge['2016년']
del cctv_pop_merge['2017년']
del cctv_pop_merge['2018년']
cctv_pop_merge.head()
```

```python
# index 설졍 함수 : set_index()
# 그래프 사용시 index값을 주기 위하여 처리함
cctv_pop_merge.set_index('구별', inplace=True)
cctv_pop_merge.head()
```

```python
# 데이터 정렬
cctv_pop_merge.sort_values(by='소계', ascending=False).head(5)
```

### 서울시 CCTV현황과 인구현황 시각화 하기

#### matplotlib 를 사용하여 그래프 시각화 처리

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import platform

from matplotlib import font_manager, rc
# import 3개
plt.rcParams['axes.unicode_minus'] = False
# unicode = false
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('System error~~~')     
```

```python
cctv_pop_merge.head()
```

```python
plt.figure() # 그린다
cctv_pop_merge['소계'].plot(kind='barh', grid=True, figsize=(10,10))
# 소계를 기준으로 / barh 막대 그래프, figszie() 막대 사이즈
plt.show()
```

```python
# 가장 큰 것부터 나열
cctv_pop_merge['소계'].sort_values().plot(kind='barh', 
                                     grid=True, figsize=(10,10))
plt.show()
```

```python
cctv_pop_merge['CCTV비율'] = cctv_pop_merge['소계'] / cctv_pop_merge['인구수'] * 100

cctv_pop_merge['CCTV비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))

plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(cctv_pop_merge['인구수'], cctv_pop_merge['소계'], s=60)
# scatter / s 원형 사이즈 크기
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
# 공학용 수치해석에 사용 :numpy
import numpy as np

fp1 = np.polyfit(cctv_pop_merge['인구수'], cctv_pop_merge['소계'], 1)
fp1
```

```python
# linespace 함수는 numpy 모듈에 포함됨
# linespace(start, stop, num) : 그래프 기리기에서 수평축의 간격 만들기에 사용
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 800000, 100)
# 범위 지정
```

```python
plt.figure(figsize=(10,10))
plt.scatter(cctv_pop_merge['인구수'], cctv_pop_merge['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g') # color = r(red),g(green),b(blue)
# fx부분을 그려서 수평축을 그림
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(cctv_pop_merge['인구수'], cctv_pop_merge['소계'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 800000, 100)

cctv_pop_merge['오차'] = np.abs(cctv_pop_merge['소계'] - f1(cctv_pop_merge['인구수']))

df_sort = cctv_pop_merge.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(cctv_pop_merge['인구수'], cctv_pop_merge['소계'], s=60)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98, 
             df_sort.index[n], fontsize=15)

plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
